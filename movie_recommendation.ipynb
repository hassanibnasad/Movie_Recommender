----IMDB Weighted Average                         
    W = (Rv+Cm)/(v+m)                 
    Where                                    
    R = (Average for the movie)
    v = (Number of votes for the movie)
    m = (Minimum votes required)
    C = (Mean vote across the whole report)

```python
R = df_cleaned['vote_average']
v = df_cleaned['vote_count']
```

```python
sn.boxplot(x=df_cleaned['vote_count'])
```

```python
C = df_cleaned['vote_average'].mean()
m = df_cleaned['vote_count'].quantile(0.85)
```

```python
m
```

```python
W = ((R*v)+(C*m))/(v+m)
```

```python
df_cleaned['weighted_average_ratings']= W
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df_cleaned[['popularity','weighted_average_ratings']])
df_normalized = scaler.transform(df_cleaned[['popularity','weighted_average_ratings']]) 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer(stop_words='english',
                       ngram_range=(1, 3))
tfidf_matrix = tfidf.fit_transform(df_cleaned['overview'])

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def hybrid_recommend(title, top_n=10, alpha=0.25, beta=0.25, gamma=0.5):
    # Index of input movie
    idx = df_cleaned[df_cleaned['original_title'] == title].index[0]
    
    # Get similarity scores
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:100]  # Skip input movie itself
    
    recommendations = []
    for i, sim_score in sim_scores:
        weighted_score = df_cleaned.iloc[i]['weighted_average_ratings']
        popularity_score = df_cleaned.iloc[i]['popularity']
        
        # Final hybrid score
        final_score = alpha * weighted_score + beta * popularity_score + gamma * sim_score
        recommendations.append((i, final_score))
    
    recommendations = sorted(recommendations, key=lambda x: x[1], reverse=True)
    top_indices = [i[0] for i in recommendations[:top_n]]
    return df_cleaned[['original_title', 'weighted_average_ratings', 'popularity']].iloc[top_indices]
```
